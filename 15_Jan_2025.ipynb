```python
import pandas as pd
# Sample renewable energy sources data
renewable_sources = ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass"]

# Sample green technology project data (for DataFrame)
data = {
    "Project": ["Solar Farm A", "Wind Turbine X", "Hydropower Y", "Solar Roof Z", "Geothermal Plant P"],
    "Technology": ["Solar", "Wind", "Hydropower", "Solar", "Geothermal"],
    "Capacity (MW)": [150, 300, 200, 50, 100],  # Megawatts
    "Cost (Million $)": [200, 400, 350, 100, 250],  # Project cost
    "Location": ["California", "Texas", "Washington", "Nevada", "Idaho"],
    "Completion Year": [2023, 2024, 2022, 2025, 2023]
}
```

```python
renewable_series = pd.Series(renewable_sources)
print("Renewable Energy Sources:")
print(renewable_series)
```

```python
projects_df = pd.DataFrame(data)
projects_df.head()
```

```python
projects_df.info()
```

```python
projects_df.describe()
```

```python
high_capacity_projects = projects_df[projects_df["Capacity (MW)"] > 100]
print("\nProjects with capacity Greater than 100 MW:")
print(high_capacity_projects)
```

```python
projects_df["Cost per MW"] = projects_df["Cost (Million $)"] / projects_df["Capacity (MW)"]
print("Dataframe with column 'Cost per MW': ")
projects_df
```

```python
total_capacity = projects_df["Capacity (MW)"].sum()
total_cost=projects_df["Cost (Million $)"].sum()
print(f"\nTotal Capacity of Renewable Projects: {total_capacity} MW")
print(f"Total cost of all projects :${total_cost} million")
```

```python
grouped_data =projects_df.groupby("Technology")["Capacity (MW)"].sum()
print("\nTotal Capacity by Renewable Technology:")
print(grouped_data)
```

```python
import pandas as pd
import numpy as np
data = {"Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
        "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
        "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]}
energy_df = pd.DataFrame(data)
print(energy_df)
print(energy_df)
```

```python
cleaned_df=energy_df.dropna()
cleaned_df.head()
```

```python
ec_mean=energy_df["Energy Consumption (MWh)"].mean()
energy_df["Energy Consumption (MWh)"].fillna(ec_mean,inplace=True)
print(energy_df)
energy_df.head()
```

```python
forward_fill_df=energy_df.fillna(method="ffill")
print(forward_fill_df)
forward_fill_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]] = scaler.fit_transform(energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]])
print(energy_df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]] = scaler.fit_transform(energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]])
print("\nData after Standardisation (Z-score Scaling)")
print(energy_df)
```

```python
energy_encoded_df = pd.get_dummies(energy_df, columns=["Energy Source"])
print("\nData after One-Hot Encoding categorical variables:")
energy_encoded_df.head()
```

Visualizing using matplotlib

```python
import matplotlib.pyplot as plt
months=['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption = [1200,1300,1100,1500,1400,1600]
plt.plot(months,energy_consumption)
plt.title('Energy Consumption over 6 months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
energy_sources =['solar','wind','hydropower','biomass']
energy_values=[1200,3400,2900,2500]
plt.bar(energy_sources,energy_values,color='green')
plt.title('Energy Consumption by Renewable Energy Source')
plt.xlabel('Energy Source')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
plt.pie(energy_values,labels=energy_sources,colors=['red','green','blue','yellow'],autopct='%1.1f%%')
plt.title("Energy Consumption Share by Source")
plt.show()
```

```python
carbon_emissions = [400,500,450,300,350,550]
plt.scatter(energy_consumption, carbon_emissions,color='red')
plt.title("Energy Consumption vs. Carbon Emissions")
plt.xlabel("Energy Consumption (MWh)")
plt.ylabel("Carbon Emissions (kg)")
plt.show()
```

```python
plt.bar(energy_sources, energy_values, color='orange', edgecolor='black', linewidth=2)
plt.title('Customized Energy Consumption by Source', fontsize=14, fontweight='bold')
plt.xlabel('Energy Source', fontsize=12)
plt.ylabel('Energy Consumption (MWh)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Energy Consumption'], loc='upper left')
plt.show()
```

Visualizing using seaborn


```python
import seaborn as sns
data = {
    "Solar": [1200, 1500, 1300],
    "Wind": [3400, 3600, 3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500, 2700, 2400]
}
df = pd.DataFrame(data)
# Create a pair plot
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Between Energy sources')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

```python
sns.violinplot(data=df)
plt.title('Violin plot of Energy Consumption Distribution')
plt.show()
```

```python
energy_values = [100,200,300,400,500]
carbon_emissions = [10,20,25,40,50]
def_red=pd.DataFrame({'Energy Consumption (MWh)':energy_values,'Carbon Emissions (kg)':carbon_emissions})
sns.regplot(x='Energy Consumption (MWh)',y='Carbon Emissions (kg)',data=def_red)
plt.title('Relationship between Energy Consumption and Carbon Emissions')
plt.show()
```
